Here we use boxplot to denote wheather there are any outliners or not within the feature columns. Here a sample of columns is taken from x1 to x10 (if outliners were not found next set of sample should be checked up untill the feature comlumns). 

Outliers are data points that deviate significantly from the rest of the dataset. They can be identified visually using boxplots, like the one you have.

In these graphs, the features like x1 and x7 are more prominent, the circles beyond the whiskers are the outliers. These are values far from the majority of data points.

```python
# Define the IQR-based outlier removal function
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].apply(lambda x: x if (x >= lower_bound) and (x <= upper_bound) else None) 
    return df

# Apply the function to all numerical columns except the first three
numerical_columns = df.columns[3:]  # Skip 'Company', 'Time', and 'Financial Distress'

for col in numerical_columns:
    df = remove_outliers_iqr(df, col)

# Drop rows with None (outliers) values
df.dropna(inplace=True)



# Visualize after outlier removal
plt.figure(figsize=(15, 8))
for i, column in enumerate(columns):
    plt.subplot(2, 5, i + 1)  # 2 rows, 5 columns for plotting
    sns.boxplot(df[column])   # df should be the outlier-removed version
    plt.title(f'Boxplot: {column}')

plt.tight_layout()
plt.show()
```

For handeling the outliners IQR rule is a common method applied. This method calculates the lower and upper bound beyond which the outliers lie. The df_filtered stores the values that are withi the outliner margin that the bounds create. This function is applied on the numerical_columns whihc stores columns beside 'Company', 'Time', and 'Financial Distress'. The above graphs show a significant improvemnet to the prior results with minimal to no outliners present.

```python
# Apply one-hot encoding for x80 
df_encoded = pd.get_dummies(df, columns=['x80'], drop_first=True)
df_encoded.head()
```

Since the categorical values are not ordinal one-hot encoding was applied to categorical variables, each unique value in a categorical column became a new feature (a new column). The column names represent each possible category, and the corresponding row will have a 1 or 0 indicating if that category applies to that particular instance in this case 'True', or 'False'.

```python
from sklearn.preprocessing import MinMaxScaler

# Separate features (X) and target variable (y)
X = df_encoded.drop(['Financial Distress'], axis=1)
y = df_encoded['Financial Distress']

# Normalize numerical features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Convert scaled features back to a DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# Apply z-score to detect outliers
z = np.abs(stats.zscore(df[numerical_columns]))

# Set a threshold to detect outliers, typically 3
outliers = (z > 3)

# Count outliers in each column
outliers_sum = np.sum(outliers, axis=0)
print(outliers_sum)

# Optional: Remove rows with outliers
df_cleaned = df[(z < 3).all(axis=1)]

# Print shape to see reduction in rows
print("Original shape:", df.shape)
print("Cleaned shape:", df_cleaned.shape)
```
